### Q.3 Write python code to compute the following Regression loss value for the given true output and predicted output by network: 
### y_true = [11, 20,19,17,10,24,23] (Target or actual Value)
### y_pred =[12,18,19.5,18,9,23,24] (Predicted value by ANN model)
### 1. MSE 
### 2. MAE 
### 3. MBE 
### 4. Huber Loss 
### 5. Epsilon Hinge Loss 
### 6. Square Epsilon Hinge Loss
### verify each loss value by analytical method

```python
y_true = [11, 20, 19, 17, 10, 24, 23]
y_pred = [12, 18, 19.5, 18, 9, 25, 24]

mseLoss = MSE(y_true, y_pred)
maeLoss = MAE(y_true, y_pred)
mbeLoss = MBE(y_true, y_pred)
huberLoss = Huber(y_true, y_pred)
ehlLoss = epsilonHingeLoss(y_true, y_pred)
sehlLoss = squaredEpsilonHingeLoss(y_true, y_pred)
```

```python
x = ["MSE", "MAE", "MBE", "Huber", "EHL", "SEHL"]
y = [mseLoss, maeLoss, mbeLoss, huberLoss, ehlLoss, sehlLoss]

plt.figure(figsize = (10, 10))
plt.bar(x, height = y)
plt.title("Regressive Loss Function comparision")
plt.show()
```
