### Q5. Let us consider the given data points (X, Y): (1, 1), (2,1), (3,2), (4,2), (5,4) and the equation of the line passing through origin Y=0.7*X - 0.1.

```python
import numpy as np 
import matplotlib.pyplot as plt 

plt.style.use("ggplot")
%matplotlib inline

X = np.array([1, 2, 3, 4, 5]) 
Y = np.array([1, 1, 2, 2, 4])

#Equation: Y = 0.7*X - 0.1  
```

### 1. Plot the graph between X and Y using Python coding

```python
# plt.figure(figsize = (10, 10))
plt.scatter(X, Y)
plt.xlabel("X")
plt.ylabel("Y") 
plt.title("Relation between X and Y")
plt.show()
```

### 2. Calculate the predicted output using above mentioned line equation using Python coding and verify the results with analytical method.

```python
def equation(X):
    return 0.7*X - 0.1
```

```python
Y_hat = equation(X)         #Predicted Output

plt.scatter(X, Y, label = "Expected Value") 
# plt.plot(X, Y)
plt.scatter(X, Y_hat, label="Predicted Value")
# plt.plot(X, Y_hat)
plt.legend()
plt.xlabel("Input (X)")
plt.ylabel("Output (Y, Y_hat)") 
plt.title("Expected vs Predicted")
plt.show()
```

### 3. Calculate the MSE value using Python coding and verify the value with the analytical method.

```python
from losses import MSE 

loss = MSE(Y, Y_hat) 
print("MSE Loss:", loss)
```

```python
plt.scatter(X, Y, label = "Expected Value") 
# plt.plot(X, Y)
# plt.scatter(X, Y_hat, label="Predicted Value")
plt.plot(X, Y_hat, label = "Regression Line", color="blue")
plt.legend()
plt.xlabel("Input (X)")
plt.ylabel("Output (Y, Y_hat)") 
plt.title("Regression Line")
plt.show()
```
