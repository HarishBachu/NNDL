### Q1 a. Write python code for plotting the following activation functions for the input x in the range of -20 to 20.

### 1. Sigmoid
### 2. TanH 
### 3. Softsign 
### 4. ReLU 
### 5. Softmax

```python
import numpy as np 
import matplotlib.pyplot as plt 
from activations import *                   # Activation functions 1 - 5 and derivatives of functions 1 - 4 are written in the activations.py file 

plt.style.use("ggplot")
%matplotlib inline
```

```python
x = np.arange(-20, 20, 0.01)

plt.figure(figsize = (20, 10)) 

ax = plt.subplot(231) 
ax.plot(x)
ax.set_title("No activation")
ax.set_ylabel("y = x")

ax = plt.subplot(232)
ax.plot(x, sigmoid(x))
ax.set_title("Sigmoid activation")
ax.set_ylabel("y = sigmoid(x)")

ax = plt.subplot(233) 
ax.plot(x, tanh(x))
ax.set_title("Tanh activation")
ax.set_ylabel("y = tanh(x)")

ax = plt.subplot(234)
ax.plot(x, softsign(x))
ax.set_title("Softsign activation")
ax.set_ylabel("y = softsign(x)")

ax = plt.subplot(235) 
ax.plot(x, ReLU(x))
ax.set_title("ReLU activation")
ax.set_ylabel("y = ReLU(x)")

ax = plt.subplot(236)
ax.plot(x, softmax(x))
ax.set_title("Softmax activation")
ax.set_ylabel("y = softmax(x)")

plt.tight_layout()
plt.suptitle("Activation Functions", fontsize = 20)
plt.subplots_adjust(top=0.92)
plt.show()
```

### Q1 b. Write python code for the derivative of the following activation functions and their plots for the input x in the range of -20 to 20. 
### 1. Sigmoid 
### 2. TanH 
### 3. Softsign 
### 4. ReLU

```python
x = np.arange(-20, 20, 0.01)

plt.figure(figsize = (15, 10)) 

ax = plt.subplot(221) 
ax.plot(x, d_sigmoid(x), label = "Derivative")
ax.plot(x, sigmoid(x), label = "Sigmoid")
ax.set_title("Sigmoid")
ax.legend()
ax.set_ylabel("y = d_sigmoid")

ax = plt.subplot(222)
ax.plot(x, d_tanh(x), label = "Derivative")
ax.plot(x, tanh(x), label = "Tanh")
ax.set_title("Tanh")
ax.legend()
ax.set_ylabel("y = d_tanh(x)")

ax = plt.subplot(223) 
ax.plot(x, d_softsign(x), label = "Derivative")
ax.plot(x, softsign(x), label = "Softsign")
ax.set_title("Softsign")
ax.legend()
ax.set_ylabel("y = d_softsign(x)")

ax = plt.subplot(224)
ax.plot(x, d_ReLU(x), label = "Derivative")
ax.plot(x, ReLU(x), label = "ReLU")
ax.set_title("ReLU")
ax.legend()
ax.set_ylabel("y = d_ReLU(x)")

plt.tight_layout()
plt.suptitle("Derivatives of Activation Functions", fontsize = 20)
plt.subplots_adjust(top=0.92)
plt.show()
```
